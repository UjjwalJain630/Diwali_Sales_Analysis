```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\Data Analysis File\Diwali Sales Data.csv", encoding = "unicode_escape")
```

```python
#Drop is a function() for delete the complete row/coloumn, axis 1 means row, 2 means coloumn, inplace true means changes in real data
df.drop(["Status","unnamed1"], axis=1, inplace= True)
```

```python
# isnull function() return boolean values false means avalable the values, true means null values
pd.isnull(df)
```

```python
#isnull with sum fuction are using for sum of all null values
pd.isnull(df).sum()
```

```python
# Dropna function are used for deleting all null values
df.dropna(inplace = True)
```

```python
#Change Data Type
df['Amount'] = df['Amount'].astype(int)
df.info()
```

```python
# rename function are used for change the coloumn name
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
# describe function are used for all discription like as min, mode, count, standard darivation
df.describe()
```

# Now we are descrie the data on the basis of STATES

```python
df.columns
```

```python
#Total number of order from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x= 'State', y='Orders')
```

# on the basis of occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

# PRODUCT CATAGORY

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x ='Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_Category', y='Amount')
```

 From the above chart we see the sold products are FOOD, CLOTHES AND ELECTRONIC CATEGORY

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_ID', y='Orders')
```

```python
# Most of Top 10 saling products 
fig1, ax1 = plt.subplots(figsize= (12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
